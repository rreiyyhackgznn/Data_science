## Simulation

```python
# Flip a coin 100 times and report the number of heads in 100 toss. 
# Repeat this experiment 7 times and record the number of heads for each

outcomes = np.array([])

for i in np.arange(7):
    outcomes = np.append(outcomes, sum(np.random.choice(coin, 100) == 'heads'))
outcomes
```

```python
# Repeat the same experiment 10,000 times and save results in outcomes
outcomes = np.array([])

for i in np.arange(10000):
    outcomes = np.append(outcomes, sum(np.random.choice(coin, 100) == 'heads'))
```

```python
outcomes
```

```python
results= pd.DataFrame({
    'Heads in 100 Tosses': outcomes})

results
```

```python
#Draw an histogram of the number of heads in 100 tosses. This is a Binomial Distribution
results.hist(bins = np.arange(25, 75), density=True, rwidth=0.85)
plt.xlabel("Heads in 100 Tosses")
plt.ylabel("Percent per unit")
plt.show()
```

```python
#Let's calculate the binomial distribution probabilities for fair coin tosses directly
import math

# n choose k
def nCr(n,r):
    return int(math.factorial(n) / (math.factorial(r) * math.factorial(n-r)))

# Calculate the probability of getting r heads in n coin tosses
def prob(n,r):
    return nCr(n,r)/(2**n)
```

```python
print('Probability of 50 heads in 100 coin tosses:', prob(100,50))
print('Probability of 40 heads in 100 coin tosses:', prob(100,40))
print('Probability of 60 heads in 100 coin tosses:', prob(100,60))
print('Probability of 30 heads in 100 coin tosses:', prob(100,30))
```
