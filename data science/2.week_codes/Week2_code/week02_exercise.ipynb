## Adding a row to a table

```python
streets = np.array(['Bancroft', 'Durant', 'Channing', 'Haste'])
```

```python
# Create a table with two columns
southside = pd.DataFrame({
    'Street': streets,
    'Blocks from Campus': np.arange(4)}
)

southside
```

```python
# Return the first row - index 0
southside.iloc[[0]]
```

```python
# return a new table with an additional row
southside.loc[4]=['Dwight', 4]
```

```python
southside
```

```python
# return  a table with an additional columns
southside['One-Way'] = ['Yes', 'Yes', 'No', 'Yes', 'Yes']
southside
```

```python
southside.iloc[:,2]
```

```python
southside.loc[:,'One-Way']
```

## Extending a table with a new column ##

```python
initial_count = minard['Survivors'][0] # first value of column Survivors
```

```python
proportion_surviving = minard['Survivors']/initial_count #Divide all values in column Survivors with the same values from this column
```

```python
proportion_surviving
```

```python
minard['Percent surviving']= proportion_surviving * 100 # add  proportion_surviving to the table
```

```python
minard
```

## Example - Working with Columns ##

```python
movies = pd.read_csv('movies_by_year_with_ticket_price.csv')
```

```python
movies
```

```python
movies.columns
```

```python
len(movies)
```

```python
movies.shape
```

```python
number_of_tix = movies['Total Gross'] * (10 ** 6) / movies['Average Ticket Price']  # calculate number of tickets sold per year by dividing gross ticket sales with average ticket price
```

```python
number_of_tix
```

```python
movies['Number of tickets']= number_of_tix # ... and add it as additional column
```

```python
movies
```

## Line Graph

```python
movies.plot('Year', 'Number of tickets')  # plot year vs. number of tickets sold
```
